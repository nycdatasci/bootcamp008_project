```python
import pipeline.pipeline as pipeline
import db.knowhere_db as kdb
from datetime import datetime
import shutil
```

```python
user = raw_input('username:')
```

```python
incoming_dir = r'../data/iPhone/incoming/'
commute_dir = r'../data/iPhone/commute/'
processed_dir = r'../data/iPhone/processed/'
clean_dir = r'../data/iPhone/clean/'
```

```python
from glob import glob
writer = kdb.Writer('knowhere')
# first normal data
for file_in in glob(incoming_dir + r'\*{}*.csv'.format(user)):
    fName = file_in.split('\\')[-1]
    fName = fName.split('.')[0] + '_clean' + '.' + fName.split('.')[1]
    file_out = processed_dir + fName
    dd = pipeline.iphone(user, file_in, file_out, commute=False)
    writer.write(data=dd, collection='iphone_stage')
    shutil.move(file_in, processed_dir)
```

```python
# then commute data
for file_in in glob(commute_dir + r'\*{}*.csv'.format(user)):
    fName = file_in.split('\\')[-1]
    fName = fName.split('.')[0] + '_clean' + '.' + fName.split('.')[1]
    file_out = processed_dir + fName
    dd = pipeline.iphone(user, file_in, file_out, commute=True)
    writer.write(data=dd, collection='iphone_stage')
    shutil.move(file_in, processed_dir)
```
