```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
```

```python
photos_train = pd.read_csv('photos_train.csv', index_col=0)
```

```python
photos_train['ratio'] = photos_train['width']/photos_train['height']
photos_train['pixels'] = photos_train['width']*photos_train['height']
```

```python
photos_train.head(5)
```

```python
x = photos_train[['ratio','r','g','b', 'pixels']]
```

```python
kmeans = KMeans(n_clusters=100, random_state=0).fit(x)
kmeans.labels_
```

```python
photos_train.groupby(['cluster', 'interest_level'])['cluster'].count()
```

```python
photos_train.to_csv('photos_train_cluster.csv', index = False)
```

```python
photos = pd.read_csv('photo_features.csv', index_col=0)
```

```python
# add new columns 
photos['ratio'] = photos['width']/photos['height']
photos['pixels'] = photos['width']*photos['height']
```

```python
x = photos[['ratio','r','g','b', 'pixels', 'brightness']]
```

```python
kmeans = KMeans(n_clusters=10, random_state=0).fit(x)
kmeans.labels_
```

```python
photos['cluster'] = kmeans.labels_
```

```python
photos[['listing_id','cluster']].to_csv('photos_cluster.csv',index=False)
```

```python
t = pd.merge(photos_train, photos[['listing_id','cluster']], how='inner', on=['listing_id'])
```
